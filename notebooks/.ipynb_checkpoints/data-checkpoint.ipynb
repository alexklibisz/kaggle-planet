```python
df = pd.read_csv('../data/train.csv')

for idx, row in df.iterrows():
    if idx > 10:
        break
    img = imread('../data/train-jpg/%s.jpg' % row['image_name'])
    img2 = np.zeros((256,256,3))
    img2[:,:,0] = img[:,:,2]
    img2[:,:,1] = img[:,:,1]
    img2[:,:,2] = img[:,:,0]
    plt.imshow(img2)
    plt.title(row['tags'].replace(' ', ', '))
    plt.show()
```
